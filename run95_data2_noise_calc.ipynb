```python
# we extract the raw waveform list
from ROOT import TFile
raw_wfms = {}
froot = TFile(datafilename)
keys = froot.GetListOfKeys()
nkeys = keys.GetEntries()
print "nkeys: ",nkeys
for i in xrange(0,nkeys):
    name = keys.At(i).GetName()
    if "wftree_" in name:
        crate = int(name.split("_")[1])
        slot  = int(name.split("_")[2])
        femch = int(name.split("_")[3])
        raw_wfms[ (crate,slot,femch) ] = pd.DataFrame( root2array( datafilename, name ) )
print "n raw waveforms: ",len(raw_wfms)
```

```python
#1. Take 50 samples after the pulse, calculate standard deviation.
def get_nsamples( wfm ):
    return len(wfm[100:150])

def calc_post50( wfm ):
    return np.std( wfm[100:150] )

rms_pulse50 = {"crate":[],"slot":[],"femch":[],"pulse50rms":[],"nsamples_p50":[]}
for ch,df in raw_wfms.items():
    df['nsamples'] = np.vectorize( get_nsamples )( df['wf'] )
    df['post50_rms'] = np.vectorize( calc_post50 )( df['wf'])
    rms_pulse50['crate'].append( ch[0] )
    rms_pulse50['slot'].append( ch[1] )
    rms_pulse50['femch'].append( ch[2] )
    rms_pulse50['pulse50rms'].append( np.mean(df['post50_rms']) )
    rms_pulse50['nsamples_p50'].append( np.mean(df['nsamples']) )
    # below can be used to draw waveform for check
    #if np.max( df['wf'].values[0]-df['ped_mean'].values[0] )<300:
    #    continue
    #plt.figure()
    #name = "pulsed [%d,%d,%d]"%sigchannel
    #plt.plot(np.arange(0,tdc*len(df['wf'].values[0][:100]), tdc), df['wf'].values[0][:100]-df['ped_mean'].values[0], 'r',color=colors[name])
    #plt.xlabel('microseconds')
    #plt.grid()
    #plt.title(name)
    #plt.show()
    #print ch
    #print df
    #break
df_pulse50 = pd.DataFrame( rms_pulse50 )
print df_pulse50
```

```python
# 2. Take 100 samples from a window of a pulse, calculate standard deviation. 
# How is this window of 100 samples defined, before or after the pulse?
def calc_pulse100( wfm ):
    return np.std( wfm[:100] )
rms_pulse100 = {"crate":[],"slot":[],"femch":[],"pulse100rms":[],"nsamples_p100":[]}
for ch,df in raw_wfms.items():
    df['pulse100'] = np.vectorize( calc_pulse100 )( df['wf'])
    rms_pulse100['crate'].append( ch[0] )
    rms_pulse100['slot'].append( ch[1] )
    rms_pulse100['femch'].append( ch[2] )
    rms_pulse100['pulse100rms'].append( np.mean(df['pulse100']) )
    rms_pulse100['nsamples_p100'].append( 100 )
df_pulse100 = pd.DataFrame( rms_pulse100 )
print df_pulse100
```

```python
# 3. Take 2,000 samples from a pulse, calculate standard deviation.
def calc_pulserms( wfm ):
    return np.std( wfm )

rms_pulse = {"crate":[],"slot":[],"femch":[],"pulserms":[],"nsamples_pulse":[]}
for ch,df in raw_wfms.items():
    df['pulserms'] = np.vectorize( calc_pulserms )( df['wf'])
    rms_pulse['crate'].append( ch[0] )
    rms_pulse['slot'].append( ch[1] )
    rms_pulse['femch'].append( ch[2] )
    rms_pulse['pulserms'].append( np.mean(df['pulserms']) )
    rms_pulse['nsamples_pulse'].append( 300 )
df_pulse = pd.DataFrame( rms_pulse )
print df_pulse
```

```python
# 4. Take 9,600 samples from an event, calculate standard deviation.
def get_index( ix ):
    return ix

# prepare empty dict
rms_event = {"crate":[],"slot":[],"femch":[],"event":[],"eventrms":[],"nsamples_evt":[]}
for ch,df in raw_wfms.items():
    #print ch
    ipulse = 0
    # an event is every time reset
    df['index'] = np.arange(len(df))
    df['event'] = np.arange(len(df),dtype=np.int)/int(5)
    evt_grouped = df.groupby(['event'])
    for evt,group in evt_grouped:
        #print ch,"event ",evt
        allwfs = group['wf'].values[0]
        #print allwfs,type(allwfs)
        #print allwfs
        for i in xrange(1,len(group)):
            #print "i=",i,type(group['wf'].ix[i])
            allwfs = np.concatenate( (allwfs, group['wf'].values[i]) )
        #print allwfs
        #print "allwfs: ",type(allwfs),len(allwfs),np.std(allwfs)
        rms_event['crate'].append( ch[0] )
        rms_event['slot'].append( ch[1] )
        rms_event['femch'].append( ch[2] )
        rms_event['event'].append( evt )
        rms_event['eventrms'].append( np.std(allwfs) )
        rms_event['nsamples_evt'].append( len(allwfs))

df_event = pd.DataFrame(rms_event)
print df_event

```

```python
# 5. Take 960,000 samples from 100 events, calculate standard deviation without average.
# we now concatenate all the waveforms per channel
# prepare empty dict
rms_channel = {"crate":[],"slot":[],"femch":[],"channelrms":[],"nsamples_ch":[]}
for ch,df in raw_wfms.items():
    #print ch
    ipulse = 0
    allwfs = df['wf'].values[0]
    #print allwfs,type(allwfs)
    #print allwfs
    for i in xrange(1,len(df)):
        #print "i=",i,type(group['wf'].ix[i])
        allwfs = np.concatenate( (allwfs, df['wf'].values[i]) )
        #print allwfs
        #print "allwfs: ",type(allwfs),len(allwfs),np.std(allwfs)
    rms_channel['crate'].append( ch[0] )
    rms_channel['slot'].append( ch[1] )
    rms_channel['femch'].append( ch[2] )
    rms_channel['channelrms'].append( np.std(allwfs) )
    rms_channel['nsamples_ch'].append( len(allwfs))

df_channel = pd.DataFrame(rms_channel)
print df_channel
```
