```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from root_numpy import root2array, root2rec, tree2rec, array2root
```

```python
run = 95
subrun1 = 44
subrun2 = 55
datafilename = 'data/pulser_run%03d_subrun%03d_%03d.root'%(run,subrun1,subrun2)
print datafilename
wf_arr = root2array( datafilename, 'wftree' )
wfdf = pd.DataFrame( wf_arr )
pl_arr = root2array( datafilename, 'ptree' )
pldf = pd.DataFrame( pl_arr )
```

```python
print wfdf.columns
#print pldf.columns
tdc = 0.500 # microseconds
print tdc
```

```python
pldf[pldf['amp']>700]
```

```python
queries = {}
colors = {}

tstart = 0
tend = 100
asics = [(6,9,0),(6,9,1),(6,9,16),(6,9,17)]
sigchannel = (6,9,0)
channellist = []
for asic in asics:
    for ch in xrange(asic[2],asic[2]+16):
        channel = (asic[0],asic[1],ch)
        if channel==sigchannel:
            continue
        queries['pickup: (%d,%d,%d)'%channel] = 'crate==%d & slot==%d & femch==%d'%channel
        colors['pickup: (%d,%d,%d)'%channel] = 'blue'
print "NChannels: ",len(channellist)
queries['pulsed [%d,%d,%d]'%sigchannel] = 'crate==%d & slot==%d & femch==%d'%sigchannel
colors['pulsed [%d,%d,%d]'%sigchannel] = 'red'
sig="pulsed [%d,%d,%d]"%sigchannel
# MB30, 

arrs = {}
ntdcs = 100
T = 100*tdc
for name,query in queries.items():
    arrs[name] = wfdf.query( query )['wf'].values[0]
    #print wfdf.query( query )['ped_mean'].values[0]
    arrs[name][:] -= wfdf.query( query )['ped_mean'].values[0]
print arrs.keys()
print 'Period: ',T,'microseconds'
```

```python
# AVERAGING
ave_arrs = {}
tlen = tend-tstart
for name,arr in arrs.items():
    ave = np.zeros( tlen )
    n = 0
    for t in xrange(tstart,len(arr),2000):
        ave[:] += arr[t:t+(tlen)]
        n += 1
    ave/=float(n)
    ave_arrs[name] = ave
```

```python

for asic in asics:
    print "======  ASIC: ",asic," ============"
    crate = asic[0]
    slot = asic[1]
    firstch = asic[2]
    for ch in xrange(firstch,firstch+16,2):
        plt.figure()
        channel = (crate,slot,ch)
        if sigchannel==channel:
            name = "pulsed [%d,%d,%d]"%channel
        else:
            name = 'pickup: (%d,%d,%d)'%channel
        #plt.plot(np.arange(0,tdc*len(arrs[name][:100]), tdc), arrs[name][:100], 'r',color=colors[name])
        plt.plot(np.arange(0,tdc*len(ave_arrs[name][:tlen]), tdc), ave_arrs[name][:tlen], 'r',color=colors[name])
        plt.xlabel('microseconds')
        plt.grid()
        plt.title(name)
        plt.show()
```

```python
# asic list
class1 = [(6,9,0),(6,9,16)] # upward, diff (7, MB30, 2 ASICs,D4)
class2 = [] # downward, diff (FT7,same MB29,2 asics,D1)
class3 = [(6,9,1),(6,9,17)] # upward, correlate (FT9, MB 36, 2 asics)
class4 = [(6,12,0),(6,12,16)]

rejected = [] 

```

```python
plt.figure()
for name,arr in arrs.items():
    if name==sig:
        continue
    plt.plot(np.arange(0,tdc*len(arr[:100]), tdc), arr[:100], 'r',color=colors[name])
plt.xlabel('microseconds')
plt.ylabel('ADC')
plt.show()


for asic in class1:
    plt.figure()
    crate = asic[0]
    slot = asic[1]
    firstch = asic[2]
    for ch in xrange(firstch,firstch+16,2):
        channel = (crate,slot,ch)
        if sigchannel==channel:
            continue
        name = 'pickup: (%d,%d,%d)'%channel
        if name==sig:
            continue
        #plt.plot(np.arange(0,tdc*len(arrs[name][:100]), tdc), arrs[name][:100], 'r',color='red')
        plt.plot(np.arange(0,tdc*len(ave_arrs[name][:tlen]), tdc), ave_arrs[name][:tlen], 'r',color='red')
    plt.xlabel('microseconds')
    plt.ylabel('ADC')
    plt.title('class 1: ASIC(%d,%d,%d)'%(asic))
    plt.show()

for asic in class3:
    plt.figure()
    crate = asic[0]
    slot = asic[1]
    firstch = asic[2]
    for ch in xrange(firstch,firstch+16,2):
        channel = (crate,slot,ch)
        if sigchannel==channel:
            continue
        name = 'pickup: (%d,%d,%d)'%channel
        if name==sig:
            continue
        #plt.plot(np.arange(0,tdc*len(arrs[name][:100]), tdc), arrs[name][:100], 'r',color='red')
        plt.plot(np.arange(0,tdc*len(ave_arrs[name][:tlen]), tdc), ave_arrs[name][:tlen], 'r',color='red')
    plt.xlabel('microseconds')
    plt.ylabel('ADC')
    plt.title('class 3: ASIC(%d,%d,%d)'%(asic))
    plt.show()
```

```python
# FFT
fft_list = {}
nhalves = {}
for name,arr in arrs.items():
    xff = np.fft.fft( arr[:100] )
    xffc = np.conjugate( xff )
    xx = np.sqrt(np.real(xff*xffc))
    fft_list[name] = xx
    nhalves[name] = int(len(xx)/2)
```

```python
f = 1/(2*tdc)/(ntdcs/2) # 2 MHz
f *= 1.0e3
print T,'microseconds'
print f,'kHz per bin'
plt.figure()
plt.plot(np.linspace(0,nhalves[sig]*f,nhalves[sig]), fft_list[sig][:int(nhalves[sig])], 'r',color=colors[sig])
plt.xlabel('freq. [kHz]')
plt.ylabel('|P(w)|')
plt.title('Power Spectrum: '+sig)
plt.grid()
plt.show()

# Set rgb-max
fftsig = fft_list[sig]
print '0-200 kHz:',np.max(fftsig[4:6]),np.argmax(fftsig[:10])
print '200-400 kHz:',np.max(fftsig[14:16]),np.argmax(fftsig[10:20])
print '400-600 kHz:',np.max(fftsig[24:26]),np.argmax(fftsig[20:30])
rmax = np.max(fftsig[4:6])
gmax = np.max(fftsig[14:16])
bmax = np.max(fftsig[24:26])
```
