```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from root_numpy import root2array, root2rec, tree2rec, array2root
```

```python
run = 95
subrun1 = 198
subrun2 = 207
datafilename = 'data/pulser_run%03d_subrun%03d_%03d.root'%(run,subrun1,subrun2)
print datafilename
wf_arr = root2array( datafilename, 'wftree' )
wfdf = pd.DataFrame( wf_arr )
pl_arr = root2array( datafilename, 'ptree' )
pldf = pd.DataFrame( pl_arr )
```

```python
print wfdf.columns
#print pldf.columns
tdc = 0.500 # microseconds
print tdc
```

```python
pldf[pldf['amp']>700]
```

```python
queries = {}
colors = {}
badlist = [(6, 12, 18), (6, 12, 28), (7, 4, 0), (9, 16, 0), (9, 16, 2), (9, 16, 6), (9, 16, 10), (9, 16, 19)]
print "N Badlist: ",len(badlist)
for badch in badlist:
    queries['pickup: (%d,%d,%d)'%badch] = 'crate==%d & slot==%d & femch==%d'%badch
    colors['pickup: (%d,%d,%d)'%badch] = 'blue'
queries['pulsed [9,16,16]'] = 'crate==9 & slot==16 & femch==16'
colors['pulsed [9,16,16]'] = 'red'
sig="pulsed [9,16,16]"
            
arrs = {}
ntdcs = 100
T = 100*tdc
for name,query in queries.items():
    arrs[name] = wfdf.query( query )['wf'].values[0]
    #print wfdf.query( query )['ped_mean'].values[0]
    arrs[name][:] -= wfdf.query( query )['ped_mean'].values[0]
print arrs.keys()
print 'Period: ',T,'microseconds'
```

```python

for name in queries.keys():
    plt.figure()
    plt.plot(np.arange(0,tdc*len(arrs[name][:100]), tdc), arrs[name][:100], 'r',color=colors[name])
    plt.xlabel('microseconds')
    plt.grid()
    plt.title(name)
    plt.show()

```

```python
class1 = [(6,12,18),(6,12,28),(9,16,0),(9,16,2),(9,16,6),(9,16,6),(9,16,10),]
class2 = []
class3 = []
class4 = [(9,16,19)]

rejected = [(7,4,0)]

```

```python
# FFT
fft_list = {}
nhalves = {}
for name,arr in arrs.items():
    xff = np.fft.fft( arr[:100] )
    xffc = np.conjugate( xff )
    xx = np.sqrt(np.real(xff*xffc))
    fft_list[name] = xx
    nhalves[name] = int(len(xx)/2)
```

```python
f = 1/(2*tdc)/(ntdcs/2) # 2 MHz
f *= 1.0e3
print T,'microseconds'
print f,'kHz per bin'
plt.figure()
plt.plot(np.linspace(0,nhalves[sig]*f,nhalves[sig]), fft_list[sig][:int(nhalves[sig])], 'r',color=colors[sig])
plt.xlabel('freq. [kHz]')
plt.ylabel('|P(w)|')
plt.title('Power Spectrum: '+sig)
plt.grid()
plt.show()

# Set rgb-max
fftsig = fft_list[sig]
print '0-200 kHz:',np.max(fftsig[4:6]),np.argmax(fftsig[:10])
print '200-400 kHz:',np.max(fftsig[14:16]),np.argmax(fftsig[10:20])
print '400-600 kHz:',np.max(fftsig[24:26]),np.argmax(fftsig[20:30])
rmax = np.max(fftsig[4:6])
gmax = np.max(fftsig[14:16])
bmax = np.max(fftsig[24:26])
```
