```python
#ex = df[ df['max_amp']>100 ]
#wfdf.query('crate==2 & slot==6 & femch==0')
#sig_arr = wfdf.query('crate==2 & slot==6 & femch==0')['wf']
# 7,4,0: 
queries = {'pulsed [2,6,0]':'crate==2 & slot==6 & femch==0',
            'neighbor [2,6,2]':'crate==2 & slot==6 & femch==2',
            'unrelated [4,6,2]':'crate==4 & slot==6 & femch==2',
            'bad (corr.) [7,4,0]':'crate==7 & slot==4 & femch==0',
            'bad (anti.) [6,9,10]':'crate==6 & slot==9 & femch==10',
            'V678 [9,11,19]':'crate==9 & slot==11 & femch==19'}
colors = {'pulsed [2,6,0]':'red',
          'neighbor [2,6,2]':'orange',
            'unrelated [4,6,2]':'black',
            'bad (corr.) [7,4,0]':'blue',
            'bad (anti.) [6,9,10]':'purple',
            'V678 [9,11,19]':'magenta'}
arrs = {}
ntdcs = 100
T = 100*tdc
for name,query in queries.items():
    arrs[name] = wfdf.query( query )['wf'].values[0]
    print wfdf.query( query )['ped_mean'].values[0]
    arrs[name][:] -= wfdf.query( query )['ped_mean'].values[0]
print arrs.keys()
print 'Period: ',T,'microseconds'
```

```python
plt.figure()
sig = 'pulsed [2,6,0]'
for name,arr in arrs.items():
    if name==sig:
        continue
    plt.plot(np.arange(0,tdc*len(arr[:100]), tdc), arr[:100], 'r',color=colors[name])
plt.xlabel('microseconds')
plt.ylabel('ADC')
plt.show()
```

```python
# FFT
fft_list = {}
nhalves = {}
for name,arr in arrs.items():
    xff = np.fft.fft( arr[:100] )
    xffc = np.conjugate( xff )
    xx = np.sqrt(np.real(xff*xffc))
    fft_list[name] = xx
    nhalves[name] = int(len(xx)/2)
```

```python
f = 1/(2*tdc)/(ntdcs/2) # 2 MHz
f *= 1.0e3
print T,'microseconds'
print f,'kHz per bin'
plt.figure()
plt.plot(np.linspace(0,nhalves[sig]*f,nhalves[sig]), fft_list[sig][:int(nhalves[sig])], 'r',color=colors[sig])
plt.xlabel('freq. [kHz]')
plt.ylabel('|P(w)|')
plt.title('Power Spectrum: '+sig)
plt.show()
```

```python
f = 1/(2*tdc)/(ntdcs/2) # 2 MHz
f *= 1.0e3
print f,'kHz'
plt.figure()
for name,fft in fft_list.items():
    if name==sig:
        continue
    nhalf = nhalves[name]
    plt.plot(np.linspace(0,nhalves[sig]*f,nhalves[sig]), fft[:nhalf], 'r',color=colors[name])
plt.xlabel('freq. [kHz]')
plt.ylabel('|P(w)|')
plt.title('Power Spectrum')
plt.show()
```
